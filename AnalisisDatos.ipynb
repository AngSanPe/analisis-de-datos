### Resumen Numérico y Visualización de los datos

```python
dfVivienda.describe()
```

```python
# Mapeo de la Region para visualización
regiones = {
    1: "Noroeste",
    2: "Noreste",
    3: "Occidente y Bajío",
    4: "Ciudad de México",
    5: "Centro Sur y Oriente",
    6: "Sur"
}

dfVivienda["RegionMapeada"] = dfVivienda["REGION"].map(regiones)

# Visualización mapeo en la BD
dfVivienda[["REGION","RegionMapeada"]]
```

```python
# Se importa libreria para graficar
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

# Se establece el color de fondo del plot
sns.set_theme(style="darkgrid")

# Se define la gráfica, de tipo barra
sns.countplot(data=dfVivienda, x="RegionMapeada", palette=["red", "blue", "green", "purple", "brown", "orange"])

# Se rotan las etiquets para evitar que se sobrepongan una con la otra
plt.xticks(rotation=45)
plt.title("Cantidad de Encuestados por Región")
plt.tight_layout()
plt.show()

```

```python
valores_frecuencia = dfVivienda['P0_4_1A'].value_counts()
plt.figure(figsize=(10, 6))
valores_frecuencia.plot(kind='bar')

# Agregar etiquetas y título
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

```python
#visualizar la correlacion de la cantidad de habitaciones con los habitantes en ella
promedio_por_grupo = dfVivienda.groupby('P0_2')['P1_1'].mean()
plt.figure(figsize=(10, 6))
promedio_por_grupo.plot(kind='bar')

# Agregar etiquetas y título
plt.xlabel('cantidad de habitaciones')
plt.ylabel('promedio de personas')
plt.show()
```

Distribucion de riquezas por region

```python
promedio_por_grupo = dfVivienda.groupby(['TLOC','P0_2'])['P1_1'].mean()
plt.figure(figsize=(10, 6))
promedio_por_grupo.plot(kind='bar')

# Agregar etiquetas y título
plt.xlabel('tamaño de localidad y promedio de habitaciones')
plt.ylabel('promedio de personas')
plt.show()

print(promedio_por_grupo)
```
