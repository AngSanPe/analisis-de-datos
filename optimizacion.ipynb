```python
import numpy as np
from scipy.optimize import minimize_scalar

# Definición de la función objetivo
def f(t):
    numerator = (125 - 50 * t + 5 * t**2) * 100
    denominator = (1 + 0.5 * (1.5 * t**-3) * 0.25)**4
    return numerator / denominator

# Transformar para maximizar
def neg_f(t):
    return -f(t)

# Intervalo de búsqueda
bounds = (0.1, 2)

# Resolver usando minimize_scalar
result = minimize_scalar(neg_f, bounds=bounds, method='bounded')

# Mostrar los resultados
t_opt = result.x
f_max = -result.fun

print(f"El valor óptimo de t es: {t_opt:.4f}")
print(f"El valor máximo de la función es: {f_max:.4f}")
```

```python
import numpy as np

def busqueda_dorada(f, A, B, tol=1e-3, max_iter=100):
    phi = (1 + np.sqrt(5)) / 2  # Razón áurea
    inv_phi = 1 / phi
    d = (B - A) * inv_phi  # Distancia inicial

    # Inicialización de puntos
    x1 = B - d
    x2 = A + d
    fx1 = f(x1)
    fx2 = f(x2)
    iteration = 0
    iteraciones = []  # Lista para registrar las iteraciones

    while abs(B - A) > tol and iteration < max_iter:
        # Registrar la iteración actual
        iteraciones.append((iteration, A, B, -fx1, -fx2))
        if fx1 < fx2:  # Buscar en el intervalo [A, x2]
            B = x2
            x2 = x1
            fx2 = fx1
            d = (B - A) * inv_phi
            x1 = B - d
            fx1 = f(x1)
        else:  # Buscar en el intervalo [x1, B]
            A = x1
            x1 = x2
            fx1 = fx2
            d = (B - A) * inv_phi
            x2 = A + d
            fx2 = f(x2)

        iteration += 1

    # Punto óptimo
    x_opt = (A + B) / 2
    f_opt = -f(x_opt)

    # Registrar la última iteración
    iteraciones.append((iteration, A, B, -fx1, -fx2))

    return x_opt, f_opt, iteraciones
```

```python
def descenso_gradiente(f, grad_f, x0, learning_rate=0.1, tol=1e-5, max_iter=100):
    x = np.array(x0, dtype=float)
    iteraciones = []  # Para almacenar los resultados de cada iteración

    for i in range(max_iter):
        grad = grad_f(x[0], x[1])
        norm_grad = np.linalg.norm(grad)  # Norma del gradiente para verificar tolerancia
        iteraciones.append((i, x[0], x[1], f(x[0], x[1]), norm_grad))

        if norm_grad < tol:  # Criterio de convergencia
            break

        # Actualizar la posición
        x = x - learning_rate * grad

    # Crear un DataFrame con las iteraciones
    columns = ["Iteración", "x1", "x2", "f(x)", "Norma Gradiente"]
    tabla = pd.DataFrame(iteraciones, columns=columns)
    return x, f(x[0], x[1]), tabla
```
