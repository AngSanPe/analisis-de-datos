```python
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from statistics import mean
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, ConfusionMatrixDisplay
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris=pd.read_csv('/content/iris (1).csv')
AX_train,AX_test,BY_train,BY_test=train_test_split(iris,iris['species'],test_size=.3,random_state=0)
```

```python
y=BY_train
X=AX_train.drop(['species'],axis=1)
AX_test=AX_test.drop(['species'],axis=1)
```

```python
nb_classifier = GaussianNB()
scores=[]
kf = KFold(n_splits=5, shuffle=False)

for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    # Entrenar el clasificador con los datos de entrenamiento
    nb_classifier.fit(X_train, y_train)
    y_pred = nb_classifier.predict(X_test)
    score = accuracy_score(y_pred, y_test)
    print("Precisión en el conjunto de prueba:", score)
    scores.append(score)

print(scores)
print('promedio =',mean(scores))
```

```python
nb_classifierM = MultinomialNB()
kf = KFold(n_splits=5, shuffle=False)
puntuacion=[]
for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    # Entrenar el clasificador con los datos de entrenamiento
    nb_classifierM.fit(X_train, y_train)
    y_pred = nb_classifierM.predict(X_test)
    score = accuracy_score(y_pred, y_test)
    print("Precisión en el conjunto de prueba:", score)
    puntuacion.append(score)

print(puntuacion)
print('promedio =',mean(puntuacion))
```

```python
emails=pd.read_csv('/content/emails.csv')
a_train,a_test,b_train,b_test=train_test_split(emails,emails['Prediction'],test_size=.3,random_state=0)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
nb_classifier=GaussianNB()
nb_classifier.fit(X,y)
y_pred_iris=nb_classifier.predict(AX_test)

print(20*"-", "Iris Report" , 20*"-", "\n")
accGN_R=accuracy_score(BY_test,y_pred_iris)
print(accGN_R)
print(classification_report(BY_test,y_pred_iris))
cm=confusion_matrix(BY_test,y_pred_iris)
disp = ConfusionMatrixDisplay(cm)
disp.plot()

nb_classifier2=GaussianNB()
nb_classifier2.fit(w,q)
y_pred_emails= nb_classifier2.predict(a_test)

print(20*"-", "Emails Report" , 20*"-", "\n")
accMNB_R=accuracy_score(b_test,y_pred_emails)
print(accMNB_R)
print(classification_report(b_test,y_pred_emails))
cm=confusion_matrix(b_test,y_pred_emails)
disp2 = ConfusionMatrixDisplay(cm)
disp2.plot()
```
